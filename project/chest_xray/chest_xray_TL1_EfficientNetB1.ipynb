## Data Augmentation

````python
#
# Data augmentation
#
# Create a data augmentation stage with horizontal flipping, rotations, zooms
#
tensorflow.get_logger().setLevel('ERROR')  # Clear warnings in data augmentation

from tensorflow import keras
data_augmentation = keras.Sequential([ #데이터 증대 layer 의 변수 = data augmentation 
  layers.RandomFlip("horizontal"),#좌우 반전 
  layers.RandomRotation(0.2), 
  layers.RandomZoom(0.2),
  layers.RandomHeight(0.2),  # Not compatible with model 높이 20% 위아래 
  layers.RandomWidth(0.2),
  # layers.Rescaling(1./255) # keep for ResNet50V2, remove for EfficientNetB0
], name ="data_augmentation")
````

````python
#
# Visualize the augmented images
#
print("="*25, 'Train dataset', "="*25)
plt.figure(figsize=(10,10))
image_idx = 5 #인덱스를 바꿔줘 새 사진 
for images, labels in train_data.take(1):    # Make a batch of images & labels
    print(labels,images.shape)
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        aug_img = data_augmentation(tensorflow.expand_dims(images[image_idx], axis=0)) #위에서 지정된 이미지 1개를 가져옴
        #가져온 이미지 1개를 4차원 배치 구조로 확장한다(expand dims) =>이미지 증대를 딥러닝시킬때 모든 tensorflow layer을 통과하는 입력은 모두 배치 이미지 구조이기에
        print(aug_img.shape)
        plt.imshow(aug_img[0][:,:,0].numpy().astype("uint8")) # : : 0 = 전체 width 전체 height 0 
        plt.title("{}".format(names[int(labels[image_idx])]))
        plt.axis("off")
    break
plt.show()

print("="*25, 'Test dataset', "="*25)
plt.figure(figsize=(10,10))
image_idx = 0
for images, labels in test_data.take(1):    # Make a batch of images & labels
    print(labels,images.shape)
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        aug_img = data_augmentation(tensorflow.expand_dims(images[image_idx], axis=0))
        print(aug_img.shape)
        plt.imshow(aug_img[0][:,:,0].numpy().astype("uint8"))
        plt.title("{}".format(names[int(labels[image_idx])]))
        plt.axis("off")
    break
plt.show()
````

# Model: Transfer learning
- ## EfficientNetB1

````python
# transfer 모델 만들기 
# Transfer learning
#
## Using the model EfficientNetB1 for the first experiment with all the layers trainable 
# base model 을 만든다. 
base_model = keras.applications.EfficientNetB1(include_top = False) #파라미터 수는 작고 효율좋음 = efficientnet 모델 
#fcn은 직접 학습하지 않는다. CNN 으로 입력영상의 특징을 찾아내는 기능만 이용 
base_model.trainable = True #훈련된 CNN LAYER에서 훈련된 파라미터를 내 데이터로 재학습시킨다. 
#FALSE = 훈련된 파라미터를 그대로 사용 
# Check the input and output of the pretraioned model
print(base_model.input.shape) #채널 3
print(base_model.output.shape) #채널 1280 

for layer_number, layer in enumerate(base_model.layers):
  print(layer_number, layer.trainable, end=", ") #layer의 trainable 속성을 알아낸다. 
````

````python
model = keras.Sequential([
  layers.Input(shape=(224,224,1),name='input_layer'),
  # layers.Rescaling(1./255), #efficientnetb1은 rescaling안하고 0~255의 정수값을 그대로 가져온다. 
  #만약 모델이 resnet50모델로 한다면? - 기말 , resnet은 grayscale을 처리하지 못한다. = gray를 resnet모델로 어떻게 학습을 시키는지 
  data_augmentation, #데이터 증대 후 
  base_model, #efficientnet 컨벌루션 된 모델로 넣는다. 
  layers.GlobalMaxPooling2D(name = "global_max"),#240x240x1 이 7x7x1280으로 , nxn에서 큰 수로 1280벡터를 만드는것 
  layers.Dense(128,activation='relu'),
  layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0001), 
              loss = 'binary_crossentropy', # 0~1두가지 구분 = binary crosentropy
              metrics = ['accuracy'])

model.summary()
````

# Building the Model

````python
# Setting up the callbacks
# Setup EarlyStopping callback to stop training if model's val_loss doesn't improve for 3 epochs
early_stopping = keras.callbacks.EarlyStopping(monitor="val_loss", # watch the val loss metric
                                                  patience=10) # if val loss decreases for 10 epochs in a row, stop training
# Creating learning rate reduction callback
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor="val_loss",  
                                                 factor=0.25, # multiply the learning rate by 0.2 (reduce by 4x)
                                                 patience=4,  # 3,4,5 #4번 돌동안 val loss가 더 작아지지 않으면 lr=0.25 줄어듬
                                                 verbose=1, # print out when learning rate goes down 
                                                 min_lr=1e-7) #중단 

## Check the summary
for no, layer in enumerate(model.layers):
  print(no, layer.trainable)
````

````python
######################################################
# Training model using augmentated data
######################################################
#
%%time
history = model.fit(train_data, 
                    epochs=100, 
                    steps_per_epoch = len(train_data), 
                    validation_data = test_data,
                    validation_steps = len(test_data), # batchSize,
                    callbacks = [early_stopping, reduce_lr])
````

````python
# base model 을 만든다. 
base_model = keras.applications.EfficientNetB1(include_top = False) #파라미터 수는 작고 효율좋음 = efficientnet 모델 
#fcn은 직접 학습하지 않는다. CNN 으로 입력영상의 특징을 찾아내는 기능만 이용 
base_model.trainable = True  
#FALSE = 훈련된 파라미터를 그대로 사용 
# Check the input and output of the pretraioned model
print(base_model.input.shape) #채널 3
print(base_model.output.shape) #채널 1280 

for layer_number, layer in enumerate(base_model.layers):
  print(layer_number, layer.trainable, end=", ") #layer의 trainable 속성을 알아낸다. 

# Fine-tune from this layer onwards
fine_tune_at = 177  # half of the whole layers

#  Fine-tuning after layer_number larger than 170
# Freeze all the layers before the `fine_tune_at` layer
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable =  False

for layer_number, layer in enumerate(base_model.layers):
    print(layer_number, layer.name, layer.trainable, end=", ")
````

````python
model = keras.Sequential([
  layers.Input(shape=(224,224,1),name='input_layer'),
  # layers.Rescaling(1./255), #efficientnetb1은 rescaling안하고 0~255의 정수값을 그대로 가져온다. 
  #만약 모델이 resnet50모델로 한다면? - 기말 , resnet은 grayscale을 처리하지 못한다. = gray를 resnet모델로 어떻게 학습을 시키는지 
  data_augmentation, #데이터 증대 후 
  base_model, #efficientnet 컨벌루션 된 모델로 넣는다. 
  layers.GlobalMaxPooling2D(name = "global_max"),#240x240x1 이 7x7x1280으로 , nxn에서 큰 수로 1280벡터를 만드는것 
  layers.Dense(128,activation='relu'),
  layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0001), 
              loss = 'binary_crossentropy', # 0~1두가지 구분 = binary crosentropy
              metrics = ['accuracy'])

model.summary()
keras.utils.plot_model(model, show_shapes=True)
````

````python
# Setting up the callbacks
# Setup EarlyStopping callback to stop training if model's val_loss doesn't improve for 3 epochs
early_stopping = keras.callbacks.EarlyStopping(monitor="val_loss", # watch the val loss metric
                                                  patience=10) # if val loss decreases for 10 epochs in a row, stop training
# Creating learning rate reduction callback
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor="val_loss",  
                                                 factor=0.25, # multiply the learning rate by 0.2 (reduce by 4x)
                                                 patience=4,  # 3,4,5 #4번 돌동안 val loss가 더 작아지지 않으면 lr=0.25 줄어듬
                                                 verbose=1, # print out when learning rate goes down 
                                                 min_lr=1e-7) #중단 
# Check the best model and save the best model
cp_callback = keras.callbacks.ModelCheckpoint(filepath="./model/chest_xray_TL_efficientNetB1_best_weights.{epoch:02d}-{val_accuracy:.3f}.hdf5", 
                              monitor='val_accuracy', verbose=0, save_best_only=True)

## Check the summary
for no, layer in enumerate(model.layers):
  print(no, layer.trainable)
````

````python
history = model.fit(train_data, 
                    epochs=100, 
                    steps_per_epoch = len(train_data), 
                    validation_data = test_data,
                    validation_steps = len(test_data), # batchSize,
                    callbacks = [cp_callback, early_stopping, reduce_lr])
````
